# Using Pipelines 

# Libraries Used

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier

from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.compose import ColumnTransformer
```

```python
df=pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.drop(columns=['Name','PassengerId','Ticket','Cabin'], inplace=True)
```

```python
df.head()
```

```python
x_train, x_test, y_train, y_test=train_test_split(df.drop(columns=['Survived']), df['Survived'], test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
transformer1=ColumnTransformer(
transformers=[
        ('TFN2',SimpleImputer(),[2]),
        ('TFN3',SimpleImputer(strategy='most_frequent'),[6]),
    ],
    remainder='passthrough'
)
```

```python
transformer=ColumnTransformer(
transformers=[
         ('TFN1',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[[1,6]]),
    ],
    remainder='passthrough'
)
```

```python
transformer2=ColumnTransformer(
transformers=[
        ('TFN4',MinMaxScaler(),slice(0,10))
    ],
    remainder='passthrough'
)
```

```python
transformer4=SelectKBest(score_func=chi2, k=8)
```

```python
trf5=DecisionTreeClassifier()
```

# Pipiline

```python
Pipe=Pipeline([
                ('trf1',transformer1),
                ('trf',transformer),
                ('trf2',transformer2),
                ('trf4',trf5)
]
             )
```

```python
pipe=make_pipeline([transformer1,transformer,transformer2,transformer4,trf5])
```

```python
Pipe.fit(x_train,y_train)
#pipe.fit(x_train,y_train)
#pipe.named_steps
```
