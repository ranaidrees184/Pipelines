# Pipe line another Example

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier

from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.compose import ColumnTransformer
```

```python
df=pd.read_csv('covid_toy.csv')
df.drop(columns=['gender','city'], inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
x_train, x_test, y_train, y_test=train_test_split(df.drop(columns='has_covid'),df['has_covid'],test_size=0.2)
```

```python
le=LabelEncoder()
y_test=le.fit_transform(y_test)
y_test
y_train=le.fit_transform(y_train)
y_train
```

```python
x_train
```

```python
transforemer2=ColumnTransformer(
    [
        ('t1',SimpleImputer(),[1]),
    ],
    remainder='passthrough'
)
```

```python
transforemer3=ColumnTransformer(
    [
        ('t1',MinMaxScaler(),[0,1]),
    ],
    remainder='passthrough'
)
```

```python
transforemer=ColumnTransformer(
    [
        ('t1',OneHotEncoder(sparse_output=False),[2]),
    ],
    remainder='passthrough'
)
```

```python
cle=DecisionTreeClassifier()
```

```python
Pipe=Pipeline(
    [
        ('t1',transforemer2),
        ('t2',transforemer3),
        ('t3',transforemer),
        ('t4',cle)
    ]
)
```

```python
pipe=make_pipeline([transforemer2,transforemer3,transforemer])
```

```python
value=Pipe.fit(x_train,y_train)
```

```python
value
```

```python
Pipe2=Pipeline(
    [
        ('t1',transforemer2),
        ('t2',transforemer3),
        ('t3',transforemer),
        ('t4',cle)
    ]
)
```

```python
Pipe2.fit(x_train,y_train)
```

```python
y_pred=Pipe2.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
